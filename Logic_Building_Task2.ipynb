Problem 1: Unique Words in a Sentence

```python

sentence = "python is easy and python is powerful"
words = sentence.lower().split()
unique = set(words)
print(f"Unique words count: {len(unique)}")
print(f"Unique words: {unique}")

```

Problem 3: Find Maximum and Minimum Values

```python

def max_min(numbers):
    print(f"List: {numbers}")
    print(f"Max: {max(numbers)}")
    print(f"Min: {min(numbers)}")

# Test (manual loop version as per hint)
def max_min_loop(numbers):
    if not numbers:
        return
    max_val, min_val = numbers[0], numbers[0]
    for num in numbers[1:]:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num
    print(f"List: {numbers}")
    print(f"Max: {max_val}")
    print(f"Min: {min_val}")

# Test
numbers = [45, 22, 89, 10, 66]
max_min_loop(numbers)
```

Problem 4: Count Products Above a Price Threshold

```python

def count_above_threshold(prices, threshold=1000):
    count = sum(1 for price in prices if price > threshold)
    print(f"Products above {threshold}: {count}")

# Test
prices = [450, 1200, 899, 1500, 300]
count_above_threshold(prices)
```

Problem 5: Calculate Attendance Percentage

```python

def attendance_percentage(attendance):
    total_days = len(attendance)
    present_days = attendance.count("P")
    percentage = (present_days / total_days) * 100 if total_days > 0 else 0
    print(f"Attendance Percentage: {percentage}")

# Test
attendance = ["P", "P", "A", "P", "P"]
attendance_percentage(attendance)
```

Problem 6: Remove Duplicate Phone Numbers

```python
def unique_phones(phone_numbers):
    unique = set(phone_numbers)
    print(f"Unique phone numbers: {unique}")

# Test
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_phones(phone_numbers)
```

Problem 7: Count Character Frequency

```python
def char_frequency(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    print(freq)

# Test (assuming input like "python")
char_frequency("python")
```

 Problem 9: Check if a Key Exists in a Dictionary

```python
def check_key(d, key):
    if key in d:
        print("Employee exists")
    else:
        print("Employee does not exist")

# Test (assuming employee dict and key="Ravi")
employees = {"Ravi": 75000}
check_key(employees, "Ravi")
```

 Problem 10: Calculate Average Marks

```python
def average_marks(marks):
    avg = sum(marks) / len(marks) if marks else 0
    print(f"Average Marks: {avg}")

# Test
marks = [70, 80, 65, 75]  # Example leading to ~72.5
average_marks(marks)
```
