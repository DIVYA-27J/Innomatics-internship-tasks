### 1. User Login Check
**Step 1**: Define credentials  
**Step 2**: Get input  
**Step 3**: Check both conditions


```python
print("=== 1. User Login Check ===")
username = "admin"
password = "1234"

# Hardcoded test cases
test_cases = [
    ("admin", "1234"),      # Success
    ("admin", "wrong"),     # Wrong password
    ("user", "1234")        # Wrong username
]

for user, pwd in test_cases:
    print(f"\nTesting: username='{user}', password='{pwd}'")
    if user == username and pwd == password:
        print("Login Successful ✓")
    else:
        print("Invalid Credentials ✗")

print("\nAll test cases passed!")

```
