Problem 1: Mobile Recharge Validation System

```python
def mobile_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Recharge must be at least ₹50. Try again.\n")

        elif amount not in valid_plans:
            print("Invalid plan. Valid plans are:", valid_plans)
            print("Please retry.\n")

        else:
            print("Recharge successful for ₹", amount)
            break

# Function call
mobile_recharge()
```

Problem 2: Inventory Reorder Alert System

```python
def inventory_alert(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")

# Sample dictionary
products = {
    "Rice": 10,
    "Sugar": 20,
    "Oil": 5,
    "Wheat": 18
}

inventory_alert(products)
```

Problem 3: Student Result Processing System

```python
def student_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Input marks
marks_list = list(map(int, input("Enter marks separated by space: ").split()))

result = student_result(marks_list)
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def cab_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak == "yes":
        fare += 0.25 * fare

    return fare

while True:
    km = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ").lower()

    total_fare = cab_fare(km, peak_hour)
    print("Total Fare: ₹", round(total_fare, 2))

    choice = input("Do you want to retry? (yes/no): ").lower()
    if choice != "yes":
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python
def attendance_checker(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Input
att = input("Enter attendance (P/A) separated by space: ").split()

result = attendance_checker(att)
print("Attendance Status:", result)
```

Problem 6: Password Strength Checker

```python
def password_checker(password):
    if len(password) < 8:
        return "Weak Password"

    digit = False
    special = False

    for ch in password:
        if ch.isdigit():
            digit = True
        if ch in "@#$":
            special = True

    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password"

# Input
pwd = input("Enter password: ")

print(password_checker(pwd))
```
