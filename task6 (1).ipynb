 Problem 1: Smart Parking Lot Management System

```python
def smart_parking_system():
    capacity = int(input("Enter Parking Capacity: "))
    logs_input = input("Enter Vehicle Logs separated by space (IN/OUT): ")
    
    vehicle_logs = logs_input.split()
    
    parked = 0
    peak_usage = 0
    
    for log in vehicle_logs:
        if log.upper() == "IN":
            parked += 1
        elif log.upper() == "OUT" and parked > 0:
            parked -= 1
        
        if parked > peak_usage:
            peak_usage = parked
    
    print("\nCurrently Parked Vehicles:", parked)
    
    if parked > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")
    
    print("Peak Parking Usage:", peak_usage)



smart_parking_system()
```

Problem 2: Online Food Delivery Time Estimator

```python
def estimate_delivery_time():
    distance = float(input("Enter Distance (km): "))
    traffic = input("Enter Traffic Level (Low/Medium/High): ")
    weather = input("Enter Weather Condition (Clear/Rainy/Stormy): ")
    
    eta = distance * 5
    
    if traffic.lower() == "high":
        eta += 15
    elif traffic.lower() == "medium":
        eta += 10
    elif traffic.lower() == "low":
        eta += 5
    
    if weather.lower() == "rainy":
        eta += 10
    elif weather.lower() == "stormy":
        eta += 20
    
    print("\nEstimated Delivery Time:", int(eta), "minutes")


estimate_delivery_time()
```

Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy_analyzer():
    total_seats = int(input("Enter Total Seats: "))
    booked = int(input("Enter Number of Booked Seats: ")) 
    occupancy = (booked / total_seats) * 100   
    print(f"\nOccupancy: {occupancy:.0f}%")  
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")
seat_occupancy_analyzer()
```

Problem 4: Cloud Server Load Classification System 

```python
def classify_server_load():
    readings_input = input("Enter CPU readings separated by space: ")
    cpu_readings = list(map(float, readings_input.split()))
    total = 0
    for reading in cpu_readings:
        total += reading   
    average = total / len(cpu_readings)   
    print(f"\nAverage CPU Load: {average:.0f}%")   
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
classify_server_load()
```

Problem 5: Smart Classroom Resource Usage Monitor 

```python
def classroom_resource_monitor():
    resources = {} 
    n = int(input("Enter number of resources: "))
    for _ in range(n):
        name = input("Enter Resource Name: ")
        hours = float(input("Enter Usage Hours: "))
        resources[name] = hours
    overused = []
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
    if overused:
        print("\nOverused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("\nOverused Resources: None")
        print("Energy Alert: No")
classroom_resource_monitor()
```

Problem 6: Online Event Registration Capacity Controller

```python
def event_registration_controller():
    capacity = int(input("Enter Event Capacity: "))
    registrations = int(input("Enter Total Registrations: "))
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    print("\nConfirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
event_registration_controller()
```
