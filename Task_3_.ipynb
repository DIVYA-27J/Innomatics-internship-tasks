 Problem 1: Find top performers (handles ties)

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
max_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == max_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
print()

print("=== SEARCH QUERY KEYWORD ANALYSIS ===\n")
```

Problem 2: Count keyword frequencies (ignore punctuation, lowercase)

```python
query = "Buy mobile phone buy phone online"

# Clean and split into words
clean_query = query.lower().replace(',', '').replace('.', '').split()
keyword_count = {}

for word in clean_query:
    keyword_count[word] = keyword_count.get(word, 0) + 1

# Show only words appearing more than once
frequent_keywords = {word: count for word, count in keyword_count.items() if count > 1}
print(frequent_keywords)
print()

print("=== SENSOR DATA VALIDATION ===\n")
```

 Problem 3: Extract even readings as (hour, value) tuples

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []

for hour, reading in enumerate(sensor_readings, start=0):
    if reading % 2 == 0:  # Even number check
        valid_readings.append((hour, reading))

print("Valid Sensor Readings (Hour, Value):")
for hour, value in valid_readings:
    print(f"({hour}, {value})")
print()

print("=== EMAIL DOMAIN USAGE ANALYSIS ===\n")
```

Problem 4: Domain counts and percentages

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}
total_emails = len(emails)

for email in emails:
    domain = email.split('@')[1]  # Extract domain after @
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate percentages
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
print()

print("=== SALES SPIKE DETECTION ===\n")
```

Problem 5: Detect sales >30% above average

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg_sales = sum(sales) / len(sales)
threshold = avg_sales * 1.3  # 30% above average

print(f"Average sales: {avg_sales:.2f}")
print(f"Spike threshold: {threshold:.2f}")

for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
print()

print("=== DUPLICATE USER ID DETECTION ===\n")
```

 Problem 6: Find and count duplicate user IDs

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_count = {}

for user_id in user_ids:
    id_count[user_id] = id_count.get(user_id, 0) + 1

# Show only duplicates (appear >1 time)
duplicates = {uid: count for uid, count in id_count.items() if count > 1}
for uid, count in duplicates.items():
    print(f"{uid} → {count} times")

print("\n" + "="*50)
print("All problems solved successfully!")
print("="*50)
```
