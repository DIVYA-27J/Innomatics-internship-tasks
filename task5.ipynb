 Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    """
    Calculates total likes and determines post status.
    """
    total_likes = 0
    
    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like
    
    # Conditional check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example Usage
analyze_post_engagement([200, 300, 250, 300])
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_data, required_level):
    """
    Calculates average rainfall and checks adequacy.
    """
    total = 0
    
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# Example Usage
rainfall_checker([70, 75, 80, 65], 70)
```

 Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_usernames(usernames):
    """
    Detects duplicate usernames using set.
    """
    unique_users = set(usernames)
    
    if len(unique_users) < len(usernames):
        result = "Yes"
    else:
        result = "No"
    
    print("Duplicate Accounts Found:", result)


# Example Usage
detect_duplicate_usernames(["john", "alice", "john", "mike"])
```

 Problem 6: Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(prices):
    """
    Filters crop prices above ₹2000.
    """
    premium = []
    
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)


# Example Usage
filter_premium_crops([1500, 2500, 1800, 3200])
```

Problem 7: System Monitoring – Application Health Checker

```python
def application_health_checker(error_count):
    """
    Evaluates system health based on error count.
    """
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", error_count)
    print("System Status:", status)


# Example Usage
application_health_checker(7)
```

Problem 8: Banking – Daily Transaction Limit Checker

```python
def transaction_limit_checker(amount):
    """
    Checks if transaction is within daily limit.
    """
    daily_limit = 50000
    
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example Usage
transaction_limit_checker(60000)
```
